## Quadratic Programming

\begin{align*}
    \min_{\substack{x}}w^Tx \quad & + x^TQx : \\
    & Ax - b \le 0 \\
    & Cx - e = 0 \\
\end{align*}

$Q = Q^T \in R^{d \cdot d} \ge 0$ is a positive semi-definite (PSD) matrix

Claim : the objective is convex in $x$

Proof : 2nd order condition of convexity

$\nabla^2(w^Tx + x^TQx) = \nabla(w + 2Qx) = 2Q \ge 0$

### PSD Matrix

[Def] A symmetric matrix $Q$ is called PSD if $x^TQx \ge 0$ holds for all $x \in R^d$

[Notation] We write $Q \ge 0$ for a PSD matrix $Q$

[Property] $Q \ge 0 \Leftrightarrow$ Eigenvalues of $Q$ are non-negative

### QP subsumes LP and LS

* QP reduces to an LP when $Q = 0$
* For Least Squares,
$\min_x\left|Ax-b\right|^2 = \min_x{x}^T(A^TA)x - 2b^TAx + b^Tb$

### Equality-constrained least-squares

* A special QP in which the closed-form solution exists
  
$min_{x \in R^d}\left|Ax-b\right|^2 : Cx - e = 0$

such that $A \in R^{m \cdot d}, b \in R^m, C \in R^{p \cdot d}, e \in R^p$.

We will assume $m \ge d, p < d, rank(C) = p, rank(\left[\begin{array}{l}{\mathbf{A}}\\ {\mathbf{C}}\end{array}\right]) = d$

The closed-form solution of this QP is

$x*$ = d-components $\{\begin{bmatrix}
2A^T{A} & C^T \\
C & 0 \\
\end{bmatrix}^{-1} \begin{bmatrix}
2A^T{b} \\
0 \\
\end{bmatrix} \}$

where d-components is an operator taking the first $d$ components\}ix}$

### Lemma 1

If $\exists (x^*, z) \in R^{d+p}$ such that $\{\begin{bmatrix}
2A^T{A} & C^T \\
C & 0 \\
\end{bmatrix} \begin{bmatrix}
x^* \\
z \\
\end{bmatrix} \} = \begin{bmatrix}
2A^T{b} \\
e \\
\end{bmatrix} \}$,

then $x^*$ must be the optimal solution, i.e., $\left|Ax-b\right|^2 \ge \left|Ax^*-b\right|^2$ holds for all $x$ satisfying $Cx = e$

[Proof]

$\left|Ax-b\right|^2 = \left|(Ax-Ax^*) + (Ax^*-b)\right|^2$

$= \left|Ax - Ax^* \right|^2 + \left|Ax^* - b \right|^2 + 2(Ax-Ax^*)^T(Ax^*-b)$

$= \left|Ax-Ax^*\right|^2 + \left|Ax^*-b\right|^2$

$\ge \left|Ax^* - b \right|^2$

Since

$2(Ax-Ax^*)^T(Ax^*-b) = 2(x-x^*)^T{A}^T(Ax^*-b)$

$= -(x-x^*)^T{C}^Tz$ since $2A^T{A}x^* - 2A^Tb = -C^Tz$

$= -(Cx - Cx^*)^Tz$

$= -(e-e)^Tz = 0$

### Lemma 2

If $rank(C) = p$ and $rank(\begin{bmatrix}
A \\
C \\
\end{bmatrix}) = d$ then $\begin{bmatrix}
2A^T{A} & C^T \\
C & 0 \\
\end{bmatrix}$ is invertible

[Proof]

Suppose $M = \begin{bmatrix}
2A^T{A} & C^T \\
C & 0 \\
\end{bmatrix}$ is not invertible.

Then, any column in $M$ is a linear combination of other columns in $M$, i.e.

$\exists \begin{bmatrix}
\bar{x} \\
\bar{z} \\
\end{bmatrix} \neq 0 : \begin{bmatrix}
2A^T{A} & C^T \\
C & 0 \\
\end{bmatrix} \begin{bmatrix}
\bar{x} \\
\bar{z} \\
\end{bmatrix} = 0$

This gives

$2A^TA\bar{x} + C^T\bar{z} = 0$

$C\bar{x} = 0$

multiplying $\bar{x}^T$ to both sides, we get

$2\bar{x}^TA^TA\bar{x} + \bar{x}^TC^T\bar{z}^T = 0 \rightarrow \begin{bmatrix}
A \\
C \\
\end{bmatrix}\bar{x} = 0 \rightarrow \bar{x} = 0 \rightarrow C^T\bar{z} = 0 \rightarrow \bar{z} = 0$

Contradiction! ($\exists \begin{bmatrix}
\bar{x} \\
\bar{z} \\
\end{bmatrix} \neq 0$)
